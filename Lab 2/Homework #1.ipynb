```python
class Animals():
    def __init__(self,name,thing):
        self.name = name
        self.thing = thing
        
    def __str__(self):
        return f"{self.name} the {self.thing}"
    
class Cow(Animals):
    def __init__(self,name = "Steve",thing = "Cow"):
        super().__init__(name,thing)
        
class Horse(Animals):
    def __init__(self,name = "Jermey",thing = "Horse"):
        super().__init__(name,thing)
        
class Pig(Animals):
    def __init__(self,name = "Topher",thing = "Pig"):
        super().__init__(name,thing)
        
class Animal_Farm():
    def __init__(self):
        self.a_list = []
        
    def list_animals(self):
        if not self.a_list:
            print("Not here")
        else:
            for animal in self.a_list:
                print(f"- {animal}")
        
    def add(self,animal): 
        self.a_list.append(animal)
    
    def remove(self,name):
        for animal in self.a_list:
            if animal.name == name:
                self.a_list.remove(animal)
                print(f"You killed {animal}, he is no longer on the farm")
                return
            print("Not here.")
Farm = Animal_Farm()

Farm.add(Cow("Steve"))
Farm.add(Pig("Topher"))
Farm.add(Horse("Jermey"))

active = True
while active:
        print("\nA. Add Cow\nB. Add Pig\nC. Add Horse\nD. Remove Animal\nE. List Animals\n")
        option = input("Please type one of the following options:")
        
        if option == "A":
            name = input("Enter the name of the Cow you would like to add.")
            Farm.add(Cow(name))
            print(f" You have added {name} the Cow to the farm")
        elif option == "B":
            name = input("Enter the name of the Pig you would like to add.")
            Farm.add(Pig(name))
            print(f"You have added {name} the Pig to the farm")
        elif option == "C":
            name = input("Enter the name of the Horse you would like to add.")
            Farm.add(Horse(name))
            print(f" You have added {name} to the farm")
        elif option == "D":
            name = input("Enter the name of the animal you would like to remove from the farm:")
            Farm.remove(name)
        elif option == "E":
            print("These animals are currently on your farm:")
            Farm.list_animals()
        else: 
            print ("Error. Please try again")

```
