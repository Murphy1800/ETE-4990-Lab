```python
import time

class Character():
    def __init__(self, HP):
        self.HP = HP
        
class Asher(Character):
    def __init__(self, HP = 100):
        super().__init__(HP)
        self.coins = 0
        self.backpack = {'Weapon' : None, 'Armor' : 0}
        self.weapon = {'sword' : -40, 'bat' : -20, 'fist' : -10}
        
    def equip(self, item):
        if item in self.weapon:
            self.backpack['Weapon'] = item
        elif item == 'armor':
            self.backpack['Armor'] += 20
            self.HP += 20
        print(f"You Equipped {item}.")
      # Using the get() function to pull from backpack and use in battle  
    def attack(self, zombie):
        wpn_used = self.backpack.get('Weapon','fist')
        dmg = self.weapon[wpn_used]
        zombie.HP += dmg
        print(f"You attacked zombie with {wpn_used} for {-dmg} damage.")
        
    def rewards(self, total):
        self.coins += total
        print(f"You earned {total} coins.")
        
class Zombie(Character):
    def __init__(self, HP, attack_dmg, attack_tmg):
        super().__init__(HP)
        self.attack_dmg = attack_dmg
        self.attack_tmg = attack_tmg
        
    def attack(self, user):
        user.HP += self.attack_dmg
        print(f"Zombie hit Asher for {-self.attack_dmg} damage. Asher is now {user.HP} health.")
        
class Big_Boy(Zombie):
    def __init__(self):
        super().__init__(HP = 80, attack_dmg = -10, attack_tmg = 5)
        
class ComZom(Zombie):
    def __init__(self):
        super().__init__(HP = 40, attack_dmg = -5, attack_tmg = 3)
        
class ItemShop():
    items = {'health elixr': 30, 'bat': 10, 'sword': 20, 'armor': 30}
    
    def buy_items(user, item):
        if user.coins >= ItemShop.items[item]:
            user.coins -= ItemShop.items[item]
            if item == 'health elixr':
                user.HP += 20
                print("You drank the forbidden elixr!. Asher has gained 20 health points")
            else:
                user.equip(item)
            print(f"You bought and equipped {item}.")
        else:
            print("You cant afford this yet. Go kill some zombies!")
            
def main_menu(asher):
    while True:
        choice = input("\nMain Menu:\n1. Battle Zombie\n2. View Inventory\n3. Open Item Shop\n4. Exit Game\nChoose an option: ")

        if choice == "1":
            battle_choice(asher)
        elif choice == "2":
            open_backpack(asher)
        elif choice == "3":
            shop_options(asher)
        elif choice == "4":
            print("Exiting Game...")
            break
        else:
            print("Option not available.")
        
def open_backpack(asher):
    print("\nInventory:")
    print(f"Weapon: {asher.backpack['Weapon'] or 'None'}")
    print(f"Armor: {'Armor Equipped' if asher.backpack['Armor'] else 'No'}")
    print(f"Total Coins: {asher.coins}")
          
def battle_choice(asher):
    choice = input("\nBattle Menu:\n1. Battle Common zombie in the Trail\n2. Battle Big Boy in the Camp\n Choose an option: ")
    
    if choice == "1":
        zombie = ComZom()
        if not battle(asher, zombie):
            print("You have been defeated in the trail.")
            return
    elif choice == "2":
        zombie = Big_Boy()
        if not battle(asher, zombie):
            print("You have been defeated in the Camp.")
            return
    else:
        print("Option not available.")
          
def shop_options(asher):
    choice = input("\nItem Shop:\n1. Bat - 10 coins\n2. Sword - 20 coins\n3. Armor - 30 coins\n4. Health Elixr - 10 coins\n Choose the item you would like to buy: ")
    items = {'1':'bat','2':'sword','3':'armor','4':'health elixr'}
    if choice in items:
          ItemShop.buy_items(asher, items[choice])
    else:
          print("Option not available.")
# Im not sure whats wrong but I cant get this battle function to work
# I feel like it has to do with my code strcuture
def battle(user, zombie):
    print("LETS FIGHT!!")
    while user.HP > 0 and zombie.HP > 0:
        user.attack(zombie)
        if zombie.HP <= 0:
            break
        time.sleep(zombie.attack_tmg)
        zombie.attack(user)
        if user.HP <= 0:
            print("Game Over.")
            return False
        time.sleep(max(3 - zombie.attack_tmg,0))
        # I found the isintance fucntion online to help with the attack intervals of the user and zombie
        # Asher will attack every 3 seconds
    if isinstance(zombie, ComZom):
          user.rewards(5)
    elif isinstance(zombie, Big_Boy):
          user.rewards(10)
    print("Asher killed the zombie")
    return True

asher = Asher()
main_menu(asher)
```
