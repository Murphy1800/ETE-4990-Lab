```python
import time
import json

class Character():
    def __init__(self, HP):
        self.HP = HP
        
class Asher(Character):
    def __init__(self, HP = 100):
        super().__init__(HP)
        self.coins = 0
        self.backpack = {'Weapon' : 'fist', 'Armor' : None, 'Charm' : None}
        self.weapon = {
            'fist': {'damage': -5 },
            'bat': {'level': 1, 'damage': [-10,-12,-14,-16,-18]},
            'spear': {'level': 1, 'damage': [-20,-22,24,26,28]},
            'sword': {'level': 1, 'damage': [-30,-32,-34,-36,-38]},
            'spells': {'level': 1, 'damage': [-40,-42,-44,-46,-48]},
            'revolver': {'level': 1, 'damage': [-50,-52,-54,-56,-58]},
            'ray gun': {'level': 1, 'damage': [-60,-62,-64,-66,-68]},
        }
        self.armor = {1:20, 2:40, 3:60, 4:80, 5:100}
        self.charm = {1:5, 2:10, 3:15, 4:20, 5: 25}
        self.purchased_items = set()#to keep tack of purchased items
        
    def equip(self, item, level=1):
        if item == 'remove':
            self.remove()
            return
        
        if item in self.weapon:
            if item in self.purchased_items or item == 'fist':
                self.backpack['Weapon'] = item #equip from weapon dictionary
                self.weapon[item]['level'] = min(max(level,1), 5)
                print(f"Asher equipped a level {self.weapon[item]['level']} {item}.")
            else:
                print("You haven't purchased this weapon yet.")
        elif item == 'armor':
            armor_level = min(max(level, 1), 5)
            self.backpack['Armor'] = self.armor[armor_level]
            self.HP += self.armor[armor_level]
            print(f"Asher Equipped level {min(max(level, 1), 5)} armor.")
        elif item == 'charm':
            charm_level = min(max(level, 1),5)
            self.backpack['Charm'] = self.charm[charm_level]
            print(f"Asher equipped level {charm_level} charm.")
        else:
            print("Item not available.")
    
    def remove(self):
        print("Type the item you would like to remove:")
        item = input()
        if item in self.backpack:
            if item == 'Armor' or item == 'Charm':
                self.HP -= self.backpack[item]
            print(f"Asher removed {item}.")
            self.backpack[item] = None
        elif item == 'Weapon':
            self.backpack['Weapon'] = 'fist'
            print("Weapon removed")
        else:
            print(f"{item} is not equipped.")
            
    def display_purchased_items(self):
        print("Purchased items: ")
        for item in self.purchased_items:
            print(item)
    
    def save_purchased_items_to_json(self, filename):
        data = {"purchased_items": list(self.purchased_items)}
        with open(filename,"w") as f:
            json.dump(data,f)
        print("Purchased items save to", filename)
      # Using the get() function to pull from backpack and use in battle  
    def attack(self, zombie):
        wpn_used = self.backpack.get('Weapon','fist')
        dmg = self.weapon.get(wpn_used, {'damage': -5})['damage']
        if wpn_used != 'fist':
            dmg = self.weapon[wpn_used]['damage'][self.weapon[wpn_used]['level']- 1]
        zombie.HP += dmg
        print(f"Asher attacked zombie with {wpn_used} for {-dmg} damage.")
        
    def rewards(self, total):
        self.coins += total
        print(f"\nYou earned {total} coins.")
        
class Zombie(Character):
    def __init__(self, HP, attack_dmg, attack_tmg):
        super().__init__(HP)
        self.attack_dmg = attack_dmg
        self.attack_tmg = attack_tmg
        
    def attack(self, user):
        user.HP += self.attack_dmg
        print(f"Zombie hit Asher for {-self.attack_dmg} damage. Asher is now {user.HP} health.")
        
class Big_Boy(Zombie):
    def __init__(self):
        super().__init__(HP = 40, attack_dmg = -10, attack_tmg = 5)
        
class ComZom(Zombie):
    def __init__(self):
        super().__init__(HP = 20, attack_dmg = -5, attack_tmg = 3)
        
class Quest_character(Character):
    def __init__(self, HP, attack_dmg):
        super().__init__(HP)
        self.attack_dmg = attack_dmg
        
class zombie_worm(Quest_character):
    def __init__(self):
        super().__init__(HP = 50 , attack_dmg = -15, attack_tmg = 5)
        
class Twin_zombies(Quest_character):
    def __init__(self):
        super().__init__(HP = 75 , attack_dmg = -25, attack_tmg = 5)
        
class Bloater(Quest_character):
    def __init__(self):
        super().__init__(HP = 100 , attack_dmg = -35, attack_tmg = 5)
        
class Horde(Quest_character):
    def __init__(self):
        super().__init__(HP = 125 , attack_dmg = -45, attack_tmg = 5)
        
class Witch(Quest_character):
    def __init__(self):
        super().__init__(HP = 150 , attack_dmg = -55, attack_tmg = 5)
        
class Giant(Quest_character):
    def __init__(self):
        super().__init__(HP = 200 , attack_dmg = -65, attack_tmg = 5)
        
class ItemShop():
    items = {'health elixr': {'price': 10, 'levels': [1]},
        'bat': {'price': [10, 12, 14, 16, 18], 'levels': [1, 2, 3, 4, 5]},
        'spear': {'price': [20, 22, 24, 26, 28], 'levels': [1, 2, 3, 4, 5]},
        'sword': {'price': [30, 32, 34, 36, 38], 'levels': [1, 2, 3, 4, 5]},
        'spells': {'price': [40, 42, 44, 46, 48], 'levels': [1, 2, 3, 4, 5]},
        'revolver': {'price': [50, 52, 54, 56, 58], 'levels': [1, 2, 3, 4, 5]},
        'ray gun': {'price': [60, 62, 64, 66, 68], 'levels': [1, 2, 3, 4, 5]},
        'armor': {'price': [25, 30, 35, 40, 45], 'levels': [1, 2, 3, 4, 5]},
        'charm': {'price': [15, 20, 25, 30, 35], 'levels': [1, 2, 3, 4, 5]}}
    @staticmethod
    def buy_items(user, item, level = 1):
        item_info = ItemShop.items[item]
        price = item_info['price'][level -1] if level in item_info['levels'] else None
        if price is not None and user.coins >= price:
            user.coins -= price
            if item == 'health elixr':
                user.HP += 20
                print("You drank the forbidden elixr! Asher has gained 20 health points")
            else:
                user.equip(item,level)
            print(f"You bought {level} {item}")
        elif price is None:
            print(f"The level {level} {item} is not available.")
        else:
            print("You can't afford this yet. Go kill some")
            
def main_menu(asher):
    common_zombies = [ComZom() for i in range(20)]
    big_boy_zombies = [Big_Boy() for i in range(10)]
    zombies_defeated = 0
    print("Every 5 zombies killed unlocks a new quest. Good Luck!")
    print(f"Kill all the zombies to win the game\nCommon Zombies: {len(common_zombies)}\nBig Boy Zombies: {len(big_boy_zombies)}")  
    while True:
        choice = input("\nMain Menu:\n1. Battle Zombie\n2. View Inventory\n3. Open Item Shop\n4. Exit Game\nChoose an option: ")

        if choice == "1":
            if battle_choice(asher,common_zombies,big_boy_zombies, zombies_defeated):
                zombies_defeated += 1
                if zombies_defeated % 5 == 0:
                    print("\nCongratulations! You have defeated 5 zombies! You've unlocked a new quest!")
                    quest_number = zombies_defeated // 5
                    new_quest(asher, quest_number)
                    if quest_number == 6:
                        print("You've vanquished all the monsters. Starting a new game...")
                        break
            if not common_zombies and not big_boy_zombies:
                print("CONGRATULATIONS! ASHER HAS DESTROYED ALL THE ZOMBIES!")
                break
        elif choice == "2":
            open_backpack(asher)
        elif choice == "3":
            shop_options(asher)
        elif choice == "4":
            print("Exiting Game...")
            break
        elif choice == "5":
            asher.display_purchased_items()
        elif choice == "6":
            filename = input("Enter the filename to save purchased items: ")
            asher.save_purchased_items_to_json(filename)
        else:
            print("Option not available.")

def new_quest(asher,quest_number):
    if quest_number == 1:
        quest_character = zombie_worm()
        print("\nYou have encountered a zombie worm for your first quest!")
        if not battle(asher,quest_character):
            print("Asher has been defeated in the quest.")
        else:
            print("Hoorah! You've completed the quest!")
            asher.rewards(30)
    elif quest_number == 2:
        quest_character = Twin_zombies()
        print("\nYou have encountered twin zombies for your second quest!")
        if not battle(asher, quest_character):
            print("Asher has been defeated in the quest.")
        else:
            print("Hoorah! You've completed the quest!")
            asher.rewards(40)
    elif quest_number == 3:
        quest_character = Bloater()
        print("\nYou have encountered a Bloater for your third quest!")
        if not battle(asher, quest_character):
            print("Asher has been defeated in the quest.")
        else:
            print("Hoorah! You've completed the quest!")
            asher.rewards(50)
    elif quest_number == 4:
        quest_character = Horde()
        print("\nYou have encountered a Horde of zombies for your fourth quest!")
        if not battle(asher, quest_character):
            print("Asher has been defeated in the quest.")
        else:
            print("Hoorah! You've completed the quest!")
            asher.rewards(60)
    elif quest_number == 5:
        quest_character = Witch()
        print("\nYou have encountered a Witch for your fifth quest!")
        if not battle(asher, quest_character):
            print("Asher has been defeated in the quest.")
        else:
            print("Hoorah! You've completed the quest!")
            asher.rewards(70)
    elif quest_number == 6:
        quest_character = Giant()
        print("\nYou have encountered a Giant for your last quest!")
        if not battle(asher, quest_character):
            print("Asher has been defeated in the quest.")
        else:
            print("You have vanquished all the monsters!")
            asher.rewards(80)
    
def open_backpack(asher):
    print("\nInventory:")
    print(f"Weapon: {asher.backpack['Weapon'] or 'None'}")
    print(f"Armor: {'Armor Equipped' if asher.backpack['Armor'] else 'No'}")
    print(f"Charm: {asher.backpack['Charm'] or 'None'}")
    print(f"Total Coins: {asher.coins}")
    print(f"Current HP: {asher.HP}")
    action = input("Choose an action (e.g. 'equip' / 'remove' or 'exit'): ")
    if action == 'equip':
        print("Type the item you would like to equip:")
        item = input()
        self.equip(item)
    elif action == 'remove':
        self.remove()
    elif action == 'exit':
        print("Returning to the main menu...")
    else:
        print("Invalid action.")
        
def battle_choice(asher, common_zombies, big_boy_zombies, zombies_defeated):
    def battle_choice(asher, common_zombies, big_boy_zombies, zombies_defeated):
    # Check if quest_num is valid
        if zombies_defeated % 5 == 0:
            quest_number = zombies_defeated // 5
        if quest_number < 1 or quest_number > 6:
            print("Invalid quest number.")
            return False
        new_quest(asher, quest_number)
    zombie = None  # Initialize zombie variable
        if common_zombies or big_boy_zombies:
        choice = input("\nBattle Menu:\n1. Battle Common zombie in the Trail\n2. Battle Big Boy in the Camp\n3. Exit to Main Menu\n Choose an option: ")
        if choice == "1":
            if common_zombies:
                zombie = common_zombies.pop(0)  # Fight each common zombie only one time
                if not battle(asher, zombie):
                    print("Asher has been defeated in the trail.")
            else:
                print("Asher has defeated all the common zombies! LEGENDARY!")
        elif choice == "2":
            if big_boy_zombies:
                zombie = big_boy_zombies.pop(0)  # Fight each big boy zombie only one time
                if not battle(asher, zombie):
                    print("Asher has been defeated in the Camp.")
            else:
                print("Asher has defeated all the Boss Zombies! IMMORTAL!")
        elif choice == "3":
            print("Exiting to Main Menu...")
            return False
        else:
            print("Invalid Choice.")
            return
    else:
        print("No zombies left to battle!")
        return False

    zombies_defeated += 1
    if zombies_defeated % 5 != 0:
        print("\nCongratulations! You have defeated 5 zombies! You've unlocked a new quest!")
        quest_number = zombies_defeated // 5
        new_quest(asher, quest_number)
        if quest_number == 6:
            print("You've vanquished all the monsters. Starting a new game...")
            return False

    if not common_zombies and not big_boy_zombies:
        print("CONGRATULATIONS! ASHER HAS DESTROYED ALL THE ZOMBIES!")
        return False


def shop_options(asher):
    while True:
        print("\nItem Shop:")
        print("1. Bat - [10, 12, 14, 16, 18] coins (levels 1-5)")
        print("2. Spear - [20, 22, 24, 26, 28] coins (levels 1-5)")
        print("3. Sword - [30, 32, 34, 36, 38] coins (levels 1-5)")
        print("4. Spells - [40, 42, 44, 46, 48] coins (levels 1-5)")
        print("5. Revolver - [50, 52, 54, 56, 58] coins (levels 1-5)")
        print("6. Ray gun - [60, 62, 64, 66, 68] coins (levels 1-5)")
        print("7. Armor - [25, 30, 35, 40, 45] coins (levels 1-5)")
        print("8. Charm - [15, 20, 25, 30, 35] coins (levels 1-5)")
        print("9. Health Elixr - 10 coins")
        print("10. Exit to main menu")
        choice = input("Choose the item you would like to buy: ")
        if choice == '10':
            print("Exiting the item shop...")
            break
        items = {'1': 'bat', '2':'spear', '3':'sword', '4':'spells','5':'revolver','6':'ray gun','7':'armor','8':'charm','9':'health elixr'}
        if choice in items:
            if items[choice] != 'health elixr':
                print("Choose the level of the item (1-5):")
                level = int(input())
                ItemShop.buy_items(asher,items[choice], level)
            else:
                ItemShop.buy_items(asher,items[choice])
        else: 
            print("Option not available")
# Im not sure whats wrong but I cant get this battle function to work
# I feel like it has to do with my code strcuture
def battle(user, quest_num):
    print("\nLETS FIGHT!!")
    while user.HP > 0:
        if quest_num == 1:
            zombie = zombie_worm()
        elif quest_num == 2:
            zombie = Twin_zombies()
        elif quest_num == 3:
            zombie = Bloater()
        elif quest_num == 4:
            zombie = Horde()
        elif quest_num == 5:
            zombie = Witch()
        elif quest_num == 6:
            zombie = Giant()
        else:
            raise ValueError("Invalid quest number")

        while user.HP > 0 and zombie.HP > 0:
            user.attack(zombie)
            if zombie.HP <= 0:
                print("The zombie has been defeated.")
                break
            time.sleep(zombie.attack_tmg)
            if user.HP <= 0:
                print("Game Over.")
                return False
            zombie.attack(user)
            if user.HP <= 0:
                print("Game Over.")
                return False
            time.sleep(max(3 - zombie.attack_tmg, 0))

        if isinstance(zombie, ComZom):
            user.rewards(5)
        elif isinstance(zombie, Big_Boy):
            user.rewards(10)
        elif isinstance(zombie, zombie_worm):
            user.rewards(25)
        elif isinstance(zombie, Twin_zombies):
            user.rewards(50)
        elif isinstance(zombie, Bloater):
            user.rewards(75)
        elif isinstance(zombie, Horde):
            user.rewards(100)
        elif isinstance(zombie, Witch):
            user.rewards(125)
        elif isinstance(zombie, Giant):
            user.rewards(150)
    return True


if __name__ == "__main__":
    asher = Asher()
    zombies_defeated = 0
    main_menu(asher)
```
